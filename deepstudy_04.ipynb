# 数据操作，数据预处理

## 一、数据操作

### 1. 初始化

```python
import torch
# x 是张量，表示一个数值组成的数组，可以是多维数组
x = torch.arange(12)
print(x)
# shap 输出张量的形状
print(x.shape)
# numel 输出张量的元素个数
print(x.numel())
# reshape 可以改变张量的形状，不改变元素数量和值
print(x.reshape(3, 4))
```

```python
import torch
# zeros 创建一个2行3列4通道的全零张量
x = torch.zeros(2, 3, 4)
print(x)
# ones 创建一个2行3列4通道的全一张量
# 2个矩阵，每个矩阵3行，每行4个元素
y = torch.ones(2, 3, 4)
print(y)
```

```python
import torch
# 为张量中的每个元素赋予确定值
x = torch.tensor([[2, 1, 4, 3],[1, 2, 3, 4],[4, 3, 2, 1]])
print(x)
```

### 2.运算

```python
import torch
# 各元素对应加减乘除，取余
x = torch.tensor([1.0, 2, 4, 8])
y = torch.tensor([2, 2, 2, 2])
print(x+y)
print(x-y)
print(x*y)
print(x/y)
print(x%y)
# **是各元素对应求幂（1.0^2）
print(x**y)
# torch.exp是以 e 为底 x 为指数的指数运算
print(torch.exp(x))
# 逻辑运算符构建张量
print(x == y)
# 对张量中所有元素进行求和，产生只有一个元素的张量
print(x.sum())
```

### 3.拼接


```python
import torch
# dtype=torch.float32 强制为浮点型
x = torch.arange(12, dtype=torch.float32).reshape((3, 4))
y = torch.tensor([[2.0, 1, 4, 3],[1, 2, 3, 4],[4, 3, 2, 1]])
# dim=0 是按行拼，dim=1 是按列拼
print(torch.cat((x, y), dim=0))
print(torch.cat((x, y), dim=1))
```

### 4.广播机制


```python
import torch
x = torch.arange(3).reshape((3, 1))
y = torch.arange(2).reshape((1, 2))
print(x)
print(y)
# 维度一样，形状不一样的，将自己复制成相同大小再相加
# x 要先复制一列，y 要复制两行
print(x+y)
```

### 5.元素增删改查

```python
import torch
x = torch.arange(12).reshape((3, 4))
# -1代表最后一个元素
print(x[-1])
# 左闭右开
print(x[1:3])
# 赋值
x[1, 2] = 999
print(x)
# :可以省略，不省更规范
x[0:2, :] = 999
print(x)
```

### 6.关于内存

```python
import torch
x = torch.arange(3).reshape((3, 1))
y = torch.arange(2).reshape((1, 2))
# id 表示的是占的内存位置
before = id(y)
y = y + x
#加完y的内存位置就变了
print(id(y) == before)
before = id(y)
# 用y += x，x[:]=x+y，对应的内存位置不会改变
y += x
print(id(y) == before)
# 原地操作, 创建一个和 y 形状一样的全零张量 z
z = torch.zeros_like(y)
print('id(z):', id(z))
z[:] = x + y;
print('id(z):', id(z))
```

### 7.张量类型转换

```python
import torch
# pytorch 是深度学习框架的核心数据结构，用于构建神经网络、自动求导、GPU 加速
x = torch.arange(3).reshape((3, 1))
# numpy 是科学计算的基础数据结构，几乎所有数据处理、数值分析库都基于它
a = x.numpy()
print(type(a))
print(type(x))
# 大小为1的张量可以转换为标量
a = torch.tensor([3.5])
print(a)
print(float(a))
print(int(a))
```

## 二、数据预处理

### 1.创建一个人工数据集，存储在csv文件中

```python
import os
# '..' 表示当前目录的上一级目录（父目录），'data' 是一个文件夹名
# ../data，也就是在父目录里创建一个叫 data 的文件夹。
os.makedirs(os.path.join('..', 'data'), exist_ok=True)
# 存储在house_tiny.csv文件中，完整路径是：../data/house_tiny.csv
data_file = os.path.join('..', 'data', 'house_tiny.csv')
# data_file 就代表了 CSV 文件的位置
# w，写入模式打开指定的文件
with open(data_file, 'w') as f:
    # 列名，房间数量，小巷，价格
    f.write('NumRooms,Alley,Price\n')
    # 每行是一个数据样本（pave 是有小巷）
    # 'NA'、'null'、'N/A' 等识别为缺失值
    f.write('NA,Pave,127500\n')
    f.write('2,NA,106000\n')
    f.write('4,NA,178100\n')
    f.write('NA,NA,140000\n')
```

### 2.从创建的csv文件中加载原始数据集

```python
import pandas as pd
# 读取文件
data = pd.read_csv(data_file)
print(data)
# 不用 print 是 html 表格形式
data
```

### 3.处理缺失数据（插值和删除）

```python
# 删除就是直接丢掉
# 插值
# data.iloc[:, 0:2] 给了 inputs，data.iloc[:, 2] 给了 outputs
inputs, outputs = data.iloc[:, 0:2], data.iloc[:, 2]
# 数值部分，缺失值用均值填充
inputs['NumRooms'] = inputs['NumRooms'].fillna(inputs['NumRooms'].mean())
print(inputs)
# 类别值或离散值部分，转换为独热编码
# True 和 False 转换为整数
inputs = pd.get_dummies(inputs, dummy_na=True, dtype=int)
print(inputs)
```

### 4.转化为张量

```python
import torch
x, y =  torch.tensor(inputs.values), torch.tensor(outputs.values)
print(x)
print(y)
```
